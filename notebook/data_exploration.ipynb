```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("../data/credit_risk_1_lakh_dataset.csv")

print("Dataset Shape:", df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python

df["credit_risk"].value_counts().plot(kind="bar", title="Credit Risk Distribution")
plt.show()
```

```python
num_cols = ["age", "income", "credit_score", "loan_amount", "existing_liabilities"]

df[num_cols].hist(figsize=(12,8))
plt.show()
```

```python
cat_cols = ["sex", "employment_status", "credit_history", "repayment_behavior"]

for col in cat_cols:
    df[col].value_counts().plot(kind="bar", title=col)
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
